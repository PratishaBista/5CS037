```python
import numpy as np
import matplotlib.pyplot as plt

# Given dataset of student scores
scores = [69, 48, 84, 58, 48, 73, 83, 48, 66, 58, 84, 66, 64, 71, 64, 66, 69, 66, 69, 71, 81, 71, 73, 69, 66, 66, 64, 58, 64, 69, 69]

# Step 1: Calculate the number of bins (classes) using square root method
n = len(scores)
class_width = int(np.sqrt(n))  # Square root of number of data points
print(f"Number of classes: {class_width}")

# Step 2: Calculate the bin edges (class intervals)
min_score = min(scores)
max_score = max(scores)
bin_edges = np.linspace(min_score, max_score, class_width + 1)  # Dividing the range into classes

# Step 3: Create the frequency table using np.histogram
frequencies, bins = np.histogram(scores, bins=bin_edges)

# Step 4: Calculate relative frequency
relative_frequencies = frequencies / n

# Step 5: Calculate cumulative frequency
cumulative_frequencies = np.cumsum(frequencies)

# Step 6: Print the Frequency Table
print("\nFrequency Table with Relative and Cumulative Frequencies:")
print("Class Interval | f | rf | Cf")
for i in range(class_width):
    print(f"{int(bins[i]):2} - {int(bins[i+1]):2} | {frequencies[i]:9} | {relative_frequencies[i]:17.4f} | {cumulative_frequencies[i]}")

# Step 7: Plot the histogram using Matplotlib
plt.hist(scores, bins=bin_edges, edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Student Scores')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given dataset of apple weights
weights = [106, 107, 123, 70, 139, 186, 111, 115, 107, 104, 107, 115, 125, 126, 119, 84, 141, 98, 81, 110, 76, 82, 109, 93, 187, 95, 111, 92, 86, 68, 130, 129, 115, 128, 100, 99, 113, 204, 90, 123, 136, 110, 131, 80, 78, 75, 118, 90, 84, 82]

# Step 1: Calculate the number of bins (classes) using square root method
n = len(weights)
class_width = int(np.sqrt(n))  # Square root of number of data points
print(f"Number of classes: {class_width}")

# Step 2: Calculate the bin edges (class intervals)
min_weight = min(weights)
max_weight = max(weights)
bin_edges = np.linspace(min_weight, max_weight, class_width + 1)  # Dividing the range into classes

# Step 3: Create the frequency table using np.histogram
frequencies, bins = np.histogram(weights, bins=bin_edges)

# Step 4: Calculate relative frequency
relative_frequencies = frequencies / n

# Step 5: Calculate cumulative frequency
cumulative_frequencies = np.cumsum(frequencies)

# Step 6: Print the Frequency Table
print("\nFrequency Table with Relative and Cumulative Frequencies:")
print("Class Interval | Frequency | Relative Frequency | Cumulative Frequency")
for i in range(class_width):
    print(f"{int(bins[i]):2} - {int(bins[i+1]):2} | {frequencies[i]:9} | {relative_frequencies[i]:17.4f} | {cumulative_frequencies[i]}")

# Step 7: Plot the histogram using Matplotlib
plt.hist(weights, bins=bin_edges, edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Weight of Apples')
plt.ylabel('Frequency')
plt.title('Histogram of Apple Weights')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
ages = [32, 40, 38, 45, 50, 34, 42, 36, 48, 55,
        39, 41, 37, 44, 49, 43, 35, 52, 46, 40,
        33, 47, 39, 51, 38, 44, 42, 36, 53, 41,
        37, 45, 39, 54, 40, 38, 36, 43, 50, 42,
        38, 55, 39, 37, 49, 46, 44, 40, 35, 52]

# Step 1: Define class intervals (bins)
min_age = min(ages)
max_age = max(ages)
class_width = 4
bins = np.arange(min_age, max_age + class_width, class_width)

# Step 2: Create histogram
plt.hist(ages, bins=bins, edgecolor='black', alpha=0.7)

# Step 3: Add labels, title, and grid
plt.title("Age Distribution of Fitness Program Participants")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 4: Display the histogram
plt.show()
```

Shape:

The histogram appears to be bimodal (two peaks) with the highest frequencies in the 36-39 and 40-43 age groups.
This suggests that the program is most popular among participants in their late 30s and early 40s.
Spread:

Ages are relatively evenly distributed, with a slight clustering in the middle age ranges.
Fewer participants are in the youngest (32-35) and oldest (52-55) age ranges.
Symmetry:

The distribution is somewhat skewed to the right, as there are fewer older participants compared to younger ones.
Gaps/Outliers:

There are no significant gaps or extreme outliers in the dataset.

```python
import matplotlib.pyplot as plt

# Data: Class intervals and corresponding frequencies
class_intervals = [(0, 40), (40, 50), (50, 60), (60, 70), (70, 100)]
frequencies = [5, 6, 8, 4, 2]
class_widths = [40, 10, 10, 10, 30]

# Calculate densities
densities = [f / width for f, width in zip(frequencies, class_widths)]

# Plotting the histogram
plt.bar([interval[0] for interval in class_intervals], densities, width=class_widths, align='edge', alpha=0.7, edgecolor='black')

# Adding labels and title
plt.title("Histogram with Density on Vertical Axis")
plt.xlabel("Weight (gms.)")
plt.ylabel("Density")
plt.xticks([interval[0] + interval[1] // 2 for interval in class_intervals], labels=[f"{interval[0]} - {interval[1]}" for interval in class_intervals])

# Display the plot
plt.show()
```

The histogram shows that the density is highest in the middle intervals, specifically for the range 50 ≤ w < 60, where the density is 0.8.
The densities for the other intervals are relatively lower, with the smallest density occurring in the last interval (70 ≤ w < 100) due to its wider class width.

The data appears to be somewhat skewed because the highest density occurs in the middle classes (50 ≤ w < 60).
The density for the first and last intervals is much lower, suggesting fewer data points in these ranges.
Insights:

Most apples fall in the 50-60g range, as this range has the highest density.
There are fewer apples in the 0-40g and 70-100g ranges, which could indicate a smaller variation in the distribution of apple weights.
