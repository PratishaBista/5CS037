```python
#Problem 4: Matrix Operations
import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A.A−1 = I
A_inverse = np.linalg.inv(A)  # Find the inverse of A
identity_matrix = np.dot(A, A_inverse)  # A * A^-1
identity_matrix = np.round(identity_matrix, decimals=10) 

# 2. Prove AB ≠ BA
AB = np.dot(A, B)  # Matrix multiplication AB
BA = np.dot(B, A)  # Matrix multiplication BA

# 3. Prove (AB)^T = B^T * A^T
AB_transpose = np.transpose(AB)  # (AB)^T
BT_AT = np.dot(np.transpose(B), np.transpose(A))  # B^T * A^T

print(
    "1. A.A^(-1) = I:\n", identity_matrix,
    "\n\n2. AB:\n", AB,
    "\nBA:\n", BA,
    "\nAB ≠ BA: ", not np.array_equal(AB, BA),
    "\n\n3. (AB)^T:\n", AB_transpose,
    "\nB^T * A^T:\n", BT_AT,
    "\n(AB)^T = B^T * A^T: ", np.array_equal(AB_transpose, BT_AT),
    sep="\n"
)

# Coefficient matrix (A) and result vector (B)
A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])
B = np.array([-1, -3, 9])

# Using the inverse method
A_inverse = np.linalg.inv(A)  # Find inverse of A
X = np.dot(A_inverse, B)  # Solution vector X = A^(-1) * B

print(
    "\n\nSolving the system of equations:",
    "\nSolution vector X:\n", X,
    sep="\n"
)

```

# How fast is NumPy? An Experiment


```python
#1. Element-wise Addition
import time

# Python Lists
list1 = [i for i in range(1_000_000)]
list2 = [i for i in range(1_000_000)]

start = time.time()

result_list_add = []
for i in range(len(list1)):
    result_list_add.append(list1[i] + list2[i])

end = time.time()
print(f"Time taken for element-wise addition using Python lists: {end - start:.5f}s")


# NumPy Arrays
array1 = np.arange(1_000_000)
array2 = np.arange(1_000_000)

start = time.time()
result_numpy_add = array1 + array2
end = time.time()
print(f"Time taken for element-wise addition using NumPy arrays: {end - start:.5f}s")
```

for element-wise addition, numpy is faster.

```python
#2. Element-wise Multiplication
import time

# Python Lists
list1 = [i for i in range(1_000_000)]
list2 = [i for i in range(1_000_000)]

start = time.time()

result_list_mult = []
for i in range(len(list1)):
    result_list_mult.append(list1[i] * list2[i])

end = time.time()
print(f"Time taken for element-wise multiplication using Python lists: {end - start:.5f}s")

# NumPy Arrays
array1 = np.arange(1_000_000)
array2 = np.arange(1_000_000)

start = time.time()
result_numpy_mult = array1 * array2
end = time.time()
print(f"Time taken for element-wise multiplication using NumPy arrays: {end - start:.5f}s")
```

numpy is faster.

```python
#3. Dot Product
import time

# Python Lists
list1 = [i for i in range(1_000_000)]
list2 = [i for i in range(1_000_000)]

start = time.time()

result_list_dot = 0
for i in range(len(list1)):
    result_list_dot += list1[i] * list2[i]

end = time.time()
print(f"Time taken for dot product using Python lists: {end - start:.5f}s")

#Numpy Arrays
array1 = np.arange(1_000_000)
array2 = np.arange(1_000_000)

start = time.time()
result_numpy_dot = np.dot(array1, array2)
end = time.time()
print(f"Time taken for dot product using NumPy arrays: {end - start:.5f}s")
```

```python
#4. Matrix Multiplication
import time

# Python Lists
matrix1 = [[j for j in range(1000)] for i in range(1000)]
matrix2 = [[j for j in range(1000)] for i in range(1000)]

start = time.time()

result_list_matmult = [[
    sum(matrix1[i][k] * matrix2[k][j] for k in range(len(matrix1[0])))
    for j in range(len(matrix2[0]))
] for i in range(len(matrix1))]

end = time.time()
print(f"Time taken for matrix multiplication using Python lists: {end - start:.5f}s")

# NumPy Arrays
matrix1_np = np.arange(1000000).reshape(1000, 1000)
matrix2_np = np.arange(1000000).reshape(1000, 1000)

start = time.time()
result_numpy_matmult = np.dot(matrix1_np, matrix2_np)
end = time.time()
print(f"Time taken for matrix multiplication using NumPy arrays: {end - start:.5f}s")
```

## From the last four cells, it has been recorded that NumPy surely is way faster than Python lists. First, I will try to list down why NumPy had the lead here. 

1. **NumPy is implemented in compiled languages that are C and Fortran.** Python, on the other hand, is an interpreted language, so NumPy is significantly faster than Python. 
2. **Python lists perform operations element by element.** In most of my code, for operations, I have implemented a loop for every element, which is inherently slower in Python. However, NumPy uses vectorized operations, i.e., it processes an entire array in one go.
